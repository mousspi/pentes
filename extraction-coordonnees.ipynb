```julia
function extract_tol_surf_prof_points(filepath::String)
    in_tol_surf_prof = false
    points = []

    open(filepath, "r") do file
        for line in eachline(file)
            line = strip(line)
            
            if startswith(line, "TOL_SURF_PROF")
                in_tol_surf_prof = true
                continue
            elseif startswith(line, "END") && in_tol_surf_prof
                in_tol_surf_prof = false
                continue
            end

            if in_tol_surf_prof && startswith(line, "PROBE")
                # Exemple ligne : PROBE = 1, 1.991, 1389.970, -450.414, -9.874, 0.042, 0.022, -0.999, 0.029
                # On extrait les valeurs apr√®s le `=`
                parts = split(line, "=")
                if length(parts) == 2
                    data = strip(parts[2])
                    values = parse.(Float64, split(data, ","))
                    push!(points, values)
                end
            end
        end
    end

    return points
end
```

```julia
points = extract_tol_surf_prof_points("M2TIR_CB2_14111_15057_final.txt")

for p in points
    println(p)
end
```
